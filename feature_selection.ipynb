Feature Selection For The Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("D:/project-pcos/pcos_prediction/PCOS_clean_data_without_infertility.xlsx", sheet_name=0)
```

```python
df.describe().T
```

```python
df.corr()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
corr_features=df.corrwith(df["PCOS (Y/N)"],method='pearson').abs().sort_values(ascending=True)
corr_features=corr_features[corr_features>0.4].index
corr_features
```

```python
df_f=df[corr_features]
df_f.head()
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(df_f.corr(), annot=True)
plt.show()
```
